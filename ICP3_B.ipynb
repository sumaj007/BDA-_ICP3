```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
X = np.linspace(0, 10, 100)                       # For line plot
x_scatter = np.random.rand(10)                   # For scatter plot
y_scatter = np.random.rand(10)

fav_foods = {'Pizza': 8, 'Burger': 5, 'Ice Cream': 7}  # For bar plot
food_names = list(fav_foods.keys())
food_prices = list(fav_foods.values())

normal_data = np.random.randn(1000)              # For histogram

# Create 2x2 subplot layout with figsize
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

# Plot 1: Line plot (Top-left)
axs[0, 0].plot(X, X / 2, color='blue')
axs[0, 0].set_title("Line Plot: X vs X/2")

# Plot 2: Scatter plot (Top-right)
axs[0, 1].scatter(x_scatter, y_scatter, color='green')
axs[0, 1].set_title("Scatter: 10 Random Points")

# Plot 3: Bar chart (Bottom-left)
axs[1, 0].bar(food_names, food_prices, color='orange')
axs[1, 0].set_title("Favourite Foods and Prices")

# Plot 4: Histogram (Bottom-right)
axs[1, 1].hist(normal_data, bins=20, color='purple')
axs[1, 1].set_title("Histogram: Normal Distribution")

# Improve spacing
plt.tight_layout()

# Show the figure
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import io

# Upload and read CSV
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]
car_sales = pd.read_csv(io.BytesIO(uploaded[filename]))

# Clean 'Price': remove $, commas, USD, spaces
car_sales['Price'] = car_sales['Price'].replace('[\$,USD\s]', '', regex=True)

# Remove final 2 digits
car_sales['Price'] = car_sales['Price'].str[:-2]

# Convert to numeric
car_sales['Price'] = pd.to_numeric(car_sales['Price'], errors='coerce')

# Add 'Total Sales' column (cumulative sum)
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add 'Sale Date' column starting from today
today = pd.to_datetime("today").normalize()  # Get today's date at midnight
car_sales['Sale Date'] = [today + timedelta(days=i) for i in range(len(car_sales))]

# View the DataFrame
print(car_sales.head())
```

```python
# Plot Sale Date vs Total Sales
car_sales.plot(x='Sale Date', y='Total Sales', kind='line', figsize=(10,5))

# Add labels and title
plt.xlabel('Sale Date')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data creation for demonstration (remove this if you have your own df)
np.random.seed(0)
df = pd.DataFrame({
    'age': np.random.randint(30, 80, 100),
    'cholesterol': np.random.randint(150, 300, 100),
    'target': np.random.randint(0, 2, 100)
})

# Filter patients over 50
df_over50 = df[df['age'] > 50]

# Create the plot
fig, ax = plt.subplots()

# Scatter plot: x=age, y=cholesterol, color by target
scatter = ax.scatter(df_over50['age'], df_over50['cholesterol'], c=df_over50['target'], cmap='viridis')

# Customize the plot
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')
ax.set_title('Age vs Cholesterol for Patients Over 50')
plt.colorbar(scatter, label='Target')

# Add mean horizontal line (mean cholesterol)
mean_chol = df_over50['cholesterol'].mean()
ax.axhline(mean_chol, color='blue', linestyle='--', linewidth=1.5, label=f'Mean Cholesterol: {mean_chol:.1f}')
ax.legend()

plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots()

scatter = ax.scatter(df_over50['age'], df_over50['cholesterol'], c=df_over50['target'], cmap='viridis')

ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')
ax.set_title('Age vs Cholesterol for Patients Over 50 (Seaborn Whitegrid Style)')
plt.colorbar(scatter, label='Target')

ax.axhline(mean_chol, color='blue', linestyle='--', linewidth=1.5, label=f'Mean Cholesterol: {mean_chol:.1f}')
ax.legend()

plt.show()
```
